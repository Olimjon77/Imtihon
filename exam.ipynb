```python

df1 = df[df['name'].notna()].groupby('fuel')['name'].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else None).reset_index()
df = pd.merge(df, df1, on='fuel', how='left', suffixes=('', '_mode'))
df['name'] = df['name'].fillna(df['name_mode'])
df = df.drop(columns=['name_mode'])
df.describe()
```

```python
df1= df[df['year'].notna()].groupby('name')['year'].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else None).reset_index()
df = pd.merge(df, df1, on='name', how='left', suffixes=('', '_mode'))
df['year'] = df['year'].fillna(df['year_mode'])
df = df.drop(columns=['year_mode'])
df.describe()

```

```python
df1= df[df['selling_price'].notna()].groupby(['name', 'year'])['selling_price'].agg('mean').reset_index()
df = pd.merge(df, df1, on=['name', 'year'], how='left', suffixes=('', '_mean'))
df['selling_price'] = df['selling_price'].fillna(df['selling_price_mean'])
df = df.drop(columns=['selling_price_mean'])
df.describe()

```

```python
df1= df[df['km_driven'].notna()].groupby(['name', 'year', "selling_price"])['km_driven'].agg("mean").reset_index()
df = pd.merge(df, df1, on=['name', 'year', 'selling_price'], how='left', suffixes=('', '_mean'))
df['km_driven'] = df['km_driven'].fillna(df['km_driven_mean'])
df = df.drop(columns=['km_driven_mean'])
df.describe()
```

```python
df1= df[df['seller_type'].notna()].groupby(['name', 'year', "selling_price"])['seller_type'].agg(pd.Series.mode).reset_index()
df = pd.merge(df, df1, on=['name', 'year', 'selling_price'], how='left', suffixes=('', '_mode'))
df['seller_type'] = df['seller_type'].fillna(df['seller_type_mode'])
df = df.drop(columns=['seller_type_mode'])
df.describe()
```

```python
df1= df[df['transmission'].notna()].groupby(['name', "selling_price"])['transmission'].agg(pd.Series.mode).reset_index()
df = pd.merge(df, df1, on=['name', 'selling_price'], how='left', suffixes=('', '_mode'))
df['transmission'] = df['transmission'].fillna(df['transmission_mode'])
df = df.drop(columns=['transmission_mode'])
df.describe()
```

```python
df["year"].unique()
```

```python
df1= df[df['owner'].notna()].groupby(['name', "selling_price"])['owner'].agg(pd.Series.mode).reset_index()
df = pd.merge(df, df1, on=['name', 'selling_price'], how='left', suffixes=('', '_mode'))
df['owner'] = df['owner'].fillna(df['owner_mode'])
df = df.drop(columns=['owner_mode'])
df.describe()
```

```python
df1= df[df['mileage(km/ltr/kg)'].notna()].groupby(['name', "selling_price"])['mileage(km/ltr/kg)'].agg('mean').reset_index()
df = pd.merge(df, df1, on=['name', 'selling_price'], how='left', suffixes=('', '_mean'))
df['mileage(km/ltr/kg)'] = df['mileage(km/ltr/kg)'].fillna(df['mileage(km/ltr/kg)_mean'])
df = df.drop(columns=['mileage(km/ltr/kg)_mean'])
df.describe()
```

```python
df1= df[df['engine'].notna()].groupby(['name', "selling_price"])['engine'].agg('mean').reset_index()
df = pd.merge(df, df1, on=['name', 'selling_price'], how='left', suffixes=('', '_mean'))
df['engine'] = df['engine'].fillna(df['engine_mean'])
df = df.drop(columns=['engine_mean'])
df.describe()
```

```python
df['max_power'] = pd.to_numeric(df['max_power'], errors = 'coerce', downcast = 'float')
df1= df[df['max_power'].notna()].groupby(['name', "selling_price"])['max_power'].agg('mean').reset_index()
df = pd.merge(df, df1, on=['name', 'selling_price'], how='left', suffixes=('', '_mean'))
df['max_power'] = df['max_power'].fillna(df['max_power_mean'])
df = df.drop(columns=['max_power_mean'])
df.describe()
```

```python
df1= df[df['seats'].notna()].groupby(['name', "year"])['seats'].agg(pd.Series.mode).reset_index()
df = pd.merge(df, df1, on=['name', 'year'], how='left', suffixes=('', '_mode'))
df['seats'] = df['seats'].fillna(df['seats_mode'])
df = df.drop(columns=['seats_mode'])
df.describe()
```
